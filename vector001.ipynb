```python
# 넘파이와 맷플로트리브를 임포트 한다.
import numpy as np
import matplotlib.pyplot as plt
import math


# 넘파이 배열을 생성한후 이를 그림으로 그려준다.
v=np.array([2,1])
plt.plot([0,v[0]],[0,v[1]],'r',linewidth=2)
plt.axis([-3,3,-3,3])
plt.grid()


plt.text(-0.5, -0.5, 'Origin', fontsize=12)
plt.text(v[0]+0.1, v[1]+0.1, 'v', fontsize=12)
plt.show()
```

```python
#두개의 벡터를 생성하고 이들을 더한 결과를 출력한다.
v1=np.array([3,4])
print("v1[0]:", v1[0] )
print("v1[1]:", v1[1] )
v2=np.array([5,1])
v3=v1+v2
print("v1:", v1)
print("v2:", v2)
print("v3:", v3)

#두개의 벡터의 내적을 계산한다.
v4=np.dot(v1,v2)
print("v4:", v4)

ab1=v1[0]**2+v1[1]**2
print("ab1:", ab1)
ab2=v2[0]**2+v2[1]**2
print("ab2:", ab2)
l1=math.sqrt(ab1)
l2=math.sqrt(ab2)
theta=np.arccos(v4/(l1*l2))
print ("theta:", theta)

cal1=l1*l2*math.cos(theta)
print("cal1:", cal1)    


#print("l1:", l1)
#print("l2:", l2)
```

```python
## plt.plot과 plt.arrow의 차이점
## plt.plot은 선을 그리는 함수이고, plt.arrow는 화살표를 그리는 함수입니다.
## plt.plot은 parameter 로 x,y 를 array-like 나 float로 받지만,
## plt.arrow는 시작점의 x,y 좌표와 dx,dy 즉 증분값을 파라미터로 받아서
## 결과를 나타냅니다.

plt.arrow(0,0,v1[0],v1[1],head_width=.2,width=.1,head_length=.2,color='r',linewidth=1)
#왜 두번째 벡터의 위치가 잘못되어 나올까?
plt.arrow(v1[0],v1[1],v2[0],v2[1],head_width=.2,width=.1,head_length=.2,color='b',linewidth=1)
#plt.arrow(v1[0],v1[1],v3[0],v3[1],head_width=.2,width=.1,head_length=.5,color='b',linewidth=2)
plt.arrow(0,0,v3[0],v3[1],head_width=.2,width=.1,head_length=.2,color='g',linewidth=2)
plt.axis([-1,v3[0]+1,-1,v3[1]+1])
plt.grid()
plt.show()
```

```python
# 두개의 벡터의 가위곱을 계산한다.
a1=np.array([5,3,4])
b1=np.array([-2,1,-1])
c1=np.cross(a1,b1)
print("c1:", c1)

# 두개의 벡터의 가위곱을 계산한다.
a2=np.array([1,1,1])
b2=np.array([2,2,2])
c2=np.cross(a2,b2)
print("c2:", c2)    
```
