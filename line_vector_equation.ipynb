```python
import numpy as np

import matplotlib.pyplot as plt

# Define two vectors
v1 = np.array([2, 1])  # First vector
v2 = np.array([1, 2])  # Second vector

# Create parameter values for t and s
t = np.linspace(-2, 2, 100)
s = np.linspace(-2, 2, 100)
T, S = np.meshgrid(t, s)

# Calculate points on the plane
X = v1[0] * T + v2[0] * S
Y = v1[1] * T + v2[1] * S

# Plot the plane
plt.figure(figsize=(8, 6))
plt.plot([0, v1[0]], [0, v1[1]], 'r-', label='Vector 1')
plt.plot([0, v2[0]], [0, v2[1]], 'b-', label='Vector 2')
plt.contour(X, Y, T, levels=10, alpha=0.5)
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vector Equation of Plane: r = tv₁ + sv₂')
plt.legend()
plt.axis('equal')
plt.show()

# Print the vector equation
print(f"Vector equation: r = t({v1[0]}, {v1[1]}) + s({v2[0]}, {v2[1]})")
```
