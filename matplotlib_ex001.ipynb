```python
import matplotlib.pyplot as plt
import numpy as np

# 랜덤 데이터 생성
np.random.seed(1)  # 재현 가능하도록 시드 고정
x = np.random.rand(50)   # 0~1 사이 난수 50개
y = np.random.rand(50)
colors = np.random.rand(50)   # 색상용 데이터
sizes = np.random.randint(50, 300, size=50)  # 점 크기

# 기본 산점도
plt.figure(figsize=(6, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='plasma',edgecolors='black')

plt.title("산점도 예제")
plt.xlabel("X 값")
plt.ylabel("Y 값")
plt.colorbar(label="색상 값")  # 색상 범례
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 랜덤 데이터 생성 (정규분포)
np.random.seed(0)
data = np.random.randn(1000)  # 평균 0, 표준편차 1, 개수 1000개

# 히스토그램 그리기
plt.figure(figsize=(8, 6))
plt.hist(
    data,
    bins=20,              # 구간 개수
    color='skyblue',      # 막대 색상
    edgecolor='red',    # 막대 테두리 색상
    alpha=0.7             # 투명도
)

# 제목, 라벨
plt.title("히스토그램 예제 (정규분포)")
plt.xlabel("값")
plt.ylabel("빈도수")
plt.grid(axis='y', alpha=0.5)  # y축에만 격자 추가

plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터
labels = ['사과', '바나나', '체리', '포도', '오렌지']
sizes = [21, 30, 20, 25,4]   # 각 항목 비율
colors = ['red', 'yellow', 'pink', 'purple','blue']
explode = (0, 0.1, 0, 0,0)   # 바나나 부분을 조금 띄움

# 파이 차트 그리기
plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,       # 특정 항목 강조
    autopct='%1.1f%%',     # 퍼센트 표시
    startangle=0,         # 시작 각도 (90도 = 위쪽부터 시작)
    shadow=True            # 그림자 효과
)

plt.title("과일 비율 파이 차트")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y1 = [0, 2, 4, 9, 16]
y2 = [0, 5, 3, 7, 4]
y3 = [8, 7, 5, 6, 2]
y4 = [4, 3, 2, 1, 5]

plt.figure(figsize=(8, 6))

# 기본 실선
plt.plot(x, y1, linestyle='-', label='실선 (-)')

# 점선
plt.plot(x, y2, linestyle='--', label='점선 (--)')

# 점-점선
plt.plot(x, y3, linestyle=':', label='점선 (:)')

# 점+대시 혼합
plt.plot(x, y4, linestyle='-.', label='대시점선 (-.)')

plt.title("matplotlib 선 모양(Linestyle) 예제")
plt.xlabel("X축")
plt.ylabel("Y축")
plt.legend()
plt.grid(True)
plt.show()
```
